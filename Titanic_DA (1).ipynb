```python
# Data Type Conversion
```

```python
df["age"]= df["age"].astype(int)
```

```python
df["sex"] = df["sex"].astype("category")
```

```python
df["embarked"] = df["embarked"].astype("category")
```

```python
df["embark_town"] = df["embark_town"].astype("category")
```

```python
df["who"] = df["who"].astype("category")
```

```python
df["alive"] = df["alive"].astype("category")
```

```python
df.dtypes
```

```python
df.rename(columns=
          {"survived":"Survived",
           "pclass": "Passenger_Class",
           "sex":"Gender"}, inplace= True)
```

```python
df.columns
```

```python

# FE - EDA
'''Survived', 'Passenger_Class', 'Gender', 'age', 'sibsp', 'parch',
       'fare', 'embarked', 'class', 'who', 'adult_male', 'embark_town',
       'alive', 'alone'''
```

```python
# FE
```

```python
df["family_size"] = df["sibsp"] + df["parch"] + 1
```

```python
df.head(1)
```

```python
df["fare_category"] = pd.qcut (df["fare"], q=4, labels = ["Low", "Medium", "High", "Very High"])
```

```python
df.head(1)
```

```python
df["Passenger_Class"].unique()
```

```python
df["class_category"] = df["Passenger_Class"].map({1: "First", 2: "Second", 3: "Third"})
```

```python
df.head(1)
```

```python
df["Is_Adult"] = np.where(df["family_size"] ==1, 1, 0)
```

```python
df["Age_Group"] = pd.cut(df["age"],
                         bins = [0, 12, 18, 35, 55, 100],
                         labels = ["child", "Teen", "Young", "Adult", "Senior"])
```
